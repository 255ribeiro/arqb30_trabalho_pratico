```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#import utm #convert utm / graus | graus / utm
import geopandas
```

```python
municipios = geopandas.read_file('IGBE_-_Municipal_-_Atual.shp')
rod = geopandas.read_file('RODOVIAS.shp')
```

```python
municipios.head()
```

```python
fig,axes = plt.subplots(figsize=(10,10))

#Plotando os municipios: Coloração cinza 
#Limites preto
#alpha - transparência
#axes - quadro o qual poderá ser alterado os valores de xmin, xmax, título e etc
municipios.plot(ax=axes,color='grey',edgecolor='black',alpha=0.5) 

#Plotando as rodovias com a cor azul
rod.plot(ax=axes,color='blue', lw=0.7)
```
